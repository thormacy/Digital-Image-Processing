```python
import cv2
import sys
from PIL import Image
dir = '/Users/shengwan/opencv/data/haarcascades/haarcascade_frontalface_alt2.xml'
image = cv2.imread('/Users/shengwan/Desktop/test.jpg')
size=image.shape
classfier = cv2.CascadeClassifier(dir)
color = (0, 255, 0)
grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faceRects = classfier.detectMultiScale(grey, scaleFactor=1.2, minNeighbors=3, minSize=(32, 32))
result = []
if len(faceRects) > 0:
    #Crop each face on original image
    for faceRect in faceRects: 
        x, y, w, h = faceRect
        cv2.rectangle(image, (x - 10, y - 10), (x + w + 10, y + h + 10), color, 3) #5控制绿色框的粗细

import matplotlib.pyplot as plt
plt.figure()
plt.imshow(image)
plt.show()
cv2.imwrite('/Users/shengwan/Desktop/detect.jpg',image)
```
